```python
# ====== Imports ====== #
from bs4 import BeautifulSoup
import pandas as pd
```

```python
# === Load html from file === #
with open("earth911-search.html", "r") as html:
    soup = BeautifulSoup(html, "html.parser")
```

```python
# === Iterate through recycle items, appending to list === #
recycle_items = []

for item in soup.select(".masterTooltip"):
    recycle_items.append(item.div.get_text())
```

```python
# === Create pandas series and save to csv
recycle_series = pd.Series(recycle_items)
recycle_series.to_csv("earth911_items_list.csv", index=False)
```
